```python
import numpy as np
import pandas as pd
import os
```

```python
#open file ,read it, and then close it
file = open("hw1_1_dat.txt")
lines = file.readlines()
file.close()
```

```python
### First make a nested dictinary

### Create a dictionary with key ’it’ and value being a dictionary,
### which is again a dictionary of Key:Value pairs where the Key gives
### the value of it and the Value is the list the unique v values under
### the same it value.

it_list = []
for line in lines:
    for word in line.split('  '):
        if word.startswith('it='):
            it_list.append(int(word.split(' ')[1]))
# it_list
# print((set(it_list)))
# it_values = len(set(it_list))
# it_values

v_list = []
for line in lines:
    for word in line.split('  '):
        if word.startswith('v='):
            v_list.append(int(word.split('v=')[1]))
# v_list

df = pd.DataFrame(columns = ["it", "v"])
df["it"]=it_list
df["v"]=v_list
# df

dic = {}
for my_it in set(it_list):
    mask = df.loc[:,"it"] == my_it
    df_mask = df.loc[mask,:]
    my_set = set(df_mask["v"])
    dic[my_it]=my_set

# dic
# {'it':dic}


### the number of unique values of it and its value: n_it
set(it_list)
n_it = len(set(it_list))
# print(n_it)

### the number of lines in the file and its value: n_lines

n_lines=0
for line in lines:
    n_lines = n_lines + 1
# print(n_lines)

{"n_lines":n_lines , "n_it": n_it , "it":dic}
```
