```python
import numpy as np
import pandas as pd
```

```python
### Writeapythoncodetoextractit,v,epoch,andUbeta**2sumfromthefile.
### Compute log10(Ubeta**2 sum). Let the code to save the extracted 
### data into a data frame with column names it, v, epoch,
### log10Ubeta2sum. (Note: log10Ubeta2sum is the log of Ubeta**2 sum in
### base 10).

file = open("hw1_1_dat.txt")
lines = file.readlines()
file.close()


it_list = []
for line in lines:
    for word in line.split('  '):
        if word.startswith('it='):
            it_list.append(int(word.split(' ')[1]))
# it_list
# print((set(it_list)))
# it_values = len(set(it_list))
# it_values

v_list = []
for line in lines:
    for word in line.split('  '):
        if word.startswith('v='):
            v_list.append(int(word.split('v=')[1]))
# v_list


epoch_list = []
for line in lines:
    for word in line.split('  '):
        if word.startswith('epoch='):
            epoch_list.append(word.split(' ')[1])            
# epoch_list


U_list = []
for line in lines:
    words = line.split('  ')
    for i, word in enumerate(words):
        if word.endswith('sum is'):
            U_list.append(float(words[i+1].split(' ')[0]))
# U_list

```

```python

df = pd.DataFrame()
df["it"]=it_list
df["v"]=v_list
df["epoch"]= epoch_list
df["U"] = U_list
df["log10Ubeta2sum"]= np.log10(df["U"].values)
del df["U"]
df
```
